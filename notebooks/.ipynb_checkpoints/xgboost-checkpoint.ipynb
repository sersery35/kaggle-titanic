```python
import numpy as np
import pandas as pd
import os
from utils import get_label, load_data, preprocess_dataframe
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = load_data()
label = get_label()
```

# Preprocess Dataset

```python

preprocessed_train_df = preprocess_dataframe(train_df)
preprocessed_test_df = preprocess_dataframe(test_df)

preprocessed_train_df.head(5)
```

```python
input_features = list(preprocessed_train_df.columns)
# remove unnecessary features
input_features.remove('Ticket')
input_features.remove('PassengerId')
input_features.remove('Name')
input_features.remove('Ticket_number')
input_features.remove('Ticket_item')
input_features.remove('SibSp')
input_features.remove('Parch')
# remove what we want to predict.
input_features.remove(label)

print(f'Input features: {input_features}')
```

```python
X = preprocessed_train_df[input_features]
y = preprocessed_train_df[label]

# we need to create a train / val split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.1, random_state=42)
X_test = preprocessed_test_df[input_features]
```

```python
# hyperparamater tuning
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

hyperparameter_grid = {
    # 'classifier__eta': [0.05, 0.1, 0.3],
    # 'classifier__gamma': [0, 0.01],
    'classifier__max_depth': [6, 8, 12],
    # 'classifier__n_estimators': [100, 300],
    'classifier__learning_rate': [0.01, 0.1],
    'classifier__early_stopping_rounds': [20],
    'classifier__eval_metric': ['auc', 'error'],
    'classifier__sampling_method': ['uniform', 'gradient_based'],
    'classifier__subsample': [0.7],
    # 'classifier__colsample_bytree': [0.8, 1],
    'classifier__lambda': [0.1, 0.3]
}

pipeline = Pipeline([('classifier', xgb.XGBClassifier(
    objective='binary:logistic',
    tree_method='gpu_hist',
    enable_categorical=True,
disable_default_eval_metric=True))])


def get_tuner(estimator, param_distributions: dict, use_random_search: bool = False):
    if use_random_search:
        return RandomizedSearchCV(
            estimator=estimator,
            param_distributions=param_distributions,
            # cv=3,
            n_iter=10,
            return_train_score=True,
            random_state=42,
        )
    else:
        return GridSearchCV(
            estimator=estimator,
            param_grid=param_distributions,
            return_train_score=True,
        )
```

```python
tuner = get_tuner(pipeline, hyperparameter_grid)
tuner.fit(X_train, y_train,
          classifier__eval_set=[(X_val, y_val)],
          )
```

```python
best_model = tuner.best_estimator_
```

```python
def print_scores(model, train_val_set: list[tuple]):
    train_score = model.score(train_val_set[0][0], train_val_set[0][1])
    val_score = model.score(train_val_set[1][0], train_val_set[1][1])
    print(f'Train Score: {train_score}\nVal score: {val_score}')
```

```python
print_scores(best_model, [(X_train, y_train), (X_val, y_val)])
```

```python
best_params = {
    'early_stopping_rounds': 20,
    'eta': 0.1,  # 0.05
    'eval_metric': 'auc',
    'gamma': 0.01,  # -> maybe we can set it to 0
    'max_depth': 12,
    'n_estimators': 100,
    'subsample': 0.7,
    'sampling_method': 'uniform',
    'lambda': 0.01,
    'alpha': 0.1,
    'colsample_bytree': 0.4,
    'seed': 2
}
pipeline = Pipeline([('classifier', xgb.XGBClassifier(
    objective='binary:logistic',
    tree_method='gpu_hist',
    enable_categorical=True, **best_params))])
```

```python
pipeline.fit(X_train, y_train, classifier__eval_set=[
             (X_train, y_train), (X_val, y_val)])
```

```python
print_scores(pipeline)
```

```python
# we need to convert categorical data to numeric in order to use categorical data
enc_features = input_features.copy()
enc_features.append(label)
enc_preprocessed_train_df = preprocessed_train_df.copy()[enc_features]
categorical_cols = []
# handle "Sex" column manually since its name will be changed
enc_preprocessed_train_df['Is_male'] = enc_preprocessed_train_df['Sex'].str.lower == 'male'
enc_features.append('Is_male')
enc_preprocessed_train_df = enc_preprocessed_train_df.drop(columns=['Sex'])
enc_features.remove('Sex')
# handle "Survived" column
# enc_preprocessed_train_df[label] = enc_preprocessed_train_df[label]
for col in enc_features:
    if enc_preprocessed_train_df[col].dtype == 'category':
        print(col, ' ', enc_preprocessed_train_df[col].dtype,
              ' : ', enc_preprocessed_train_df[col][0])

        dummy_df = pd.get_dummies(enc_preprocessed_train_df[col], prefix=col)
        enc_preprocessed_train_df = enc_preprocessed_train_df.drop(columns=[
                                                                   col])
        for dummy_col in dummy_df.columns:
            enc_preprocessed_train_df[dummy_col] = dummy_df[dummy_col].astype(
                'int')
    elif enc_preprocessed_train_df[col].dtype == 'bool':
        enc_preprocessed_train_df[col] = enc_preprocessed_train_df[col].astype(
            'int')
```

```python
enc_preprocessed_train_df.head()
```

```python
exact_X = enc_preprocessed_train_df.drop(columns=[label])
exact_y = enc_preprocessed_train_df[label]

# we need to create a train / val split
exact_X_train, exact_X_val, exact_y_train, exact_y_val = train_test_split(
    exact_X, exact_y, test_size=0.3, random_state=42)
```

```python
exact_tree_params = {
    'early_stopping_rounds': [20],
    'eta':[ 0.03],  # 0.05
    'eval_metric': 'error',
        'lambda': [0.1],
    'max_depth': [9],
    'n_estimators': [100],
    'subsample': [0.7],
    'sampling_method': ['uniform', 'gradient_based'],
    
    'colsample_bytree': [0.6, 1],

}
exact_tree_method_pipeline = Pipeline([('classifier', xgb.XGBClassifier(
    objective='binary:logistic',
    tree_method='exact',
    # **exact_tree_params
))])
```

```python
exact_tuner = get_tuner(exact_tree_method_pipeline, exact_tree_params)
```

```python
exact_tuner.fit(exact_X_train, exact_y_train, classifier__eval_set=[(exact_X_val, exact_y_val)])
```

```python
print_scores(exact_tuner.best_estimator_)
```
